## Activity 1: Data Loading and Exploration

### Activity 1: Data Preparation

```python
# Import necessary libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
# Importing the csv file from the provided URL
# Imported CSV file is too big, using chunking to prevent timeout issues
# Importing necessary libraries

url = "https://lms.educlaas.com/assets/courseware/v1/e206258dd748e86b8bb35fc039e90fa8/asset-v1:EDUCLaSS+PYD+PDDS-PYD-0425-24Jun2025E+type@asset+block/owid-covid-data.csv"
filename = "owid-covid-data.csv"

# Stream download in chunks
with requests.get(url, stream=True) as r:
    r.raise_for_status()
    with open(filename, 'wb') as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)

# Setting display options for better readability
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.float_format', '{:.2f}'.format)  # Format floats to 2 decimal places
```

### Display the first 5 rows of the DataFrame.

```python
df = pd.read_csv(filename)
df.head()
# Display the first 5 rows
```

### Display the last 5 rows of the DataFrame.

```python
# Display last 5 rows
df.tail()
```

### Check for missing values in the DataFrame.

```python
# Check missing values
missing_counts = df.isnull().sum()
missing_counts[missing_counts > 0].sort_values(ascending=False)
```

### Strategy: Drop columns with more than 90% missing values

```python
# Drop columns with more than 90% missing values
threshold = len(df) * 0.9
df_cleaned = df.loc[:, df.isnull().sum() < threshold]

#Checking maximum missing percentage in any column
missing_percent = df_cleaned.isnull().mean() * 100
print(f"Max missing percentage in any column: {missing_percent.max():.2f}%")

# Find dropped columns
dropped_cols = set(df.columns) - set(df_cleaned.columns)
print(f"Dropped columns: {sorted(dropped_cols)}")
```

### Convert 'date' column to datetime format

```python
# Convert 'date' column to datetime

# Drop columns and make a copy.
# Avoids SettingWithCopyWarning by using .copy() method.

df_cleaned = df.loc[:, df.isnull().sum() < threshold].copy()

# Forward fill missing values
df_cleaned = df_cleaned.ffill()

# Convert 'date' column
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])

# Extract year and month from 'date' column
df_cleaned['year'] = df_cleaned['date'].dt.year
df_cleaned['month'] = df_cleaned['date'].dt.month

# Display the first 5 rows and summary statistics of the cleaned DataFrame
from IPython.display import display
display(df_cleaned.head())

# To understand stat range and spot outliers
display(df_cleaned.describe())
```

### Impute missing values in the dataset columns

```python
# Missing values in the dataset were imputed using forward fill (ffill) and backward fill (bfill). This approach enables continuous analysis and visualization but may introduce artificial continuity where data was originally missing. The impact of imputation should be considered when interpreting results, especially for columns where real values might differ from the filled values.

df_cleaned.ffill(inplace=True)
df_cleaned.bfill(inplace=True)

# Check for any remaining missing values
print("Missing values after forward fill:\n", df_cleaned.isnull().sum())

print("Any missing values left?", df_cleaned.isnull().values.any())

```

### Visualize the WHO Regions (Continent column in dataset) with total COVID-19 cases and total deaths by using bar plots

```python
# Using Log scale for better visualization of large numbers
# Group by continent and count unique countries
# Initially Africa and Asia have the same number count for cases amd deaths, the data displayed is wrong.This is because the data is aggregated and does not represent individual countries.

# List of locations to exclude from the analysis
# These are aggregate rows that do not represent individual countries
exclude_locs = [
    'World', 'Africa', 'Asia', 'Europe', 'European Union',
    'North America', 'Oceania', 'South America', 'International'
]

# 1. Exclude aggregate rows
df_countries = df_cleaned[~df_cleaned['location'].isin(exclude_locs)]

# 2. Get the latest data for each country
latest_per_country = (
    df_countries.sort_values('date')
    .groupby(['location', 'continent'])
    .last()
    .reset_index()
)

# 3. Group by continent and sum total cases and deaths
continent_summary = latest_per_country.groupby('continent')[['total_cases', 'total_deaths']].sum()

# 4. Plot with log scale and formatted ticks + bar labels
ax = continent_summary.plot(
    kind='bar',
    figsize=(10, 6),
    title='Total Cases and Deaths by Continent (Summed, No Aggregates)'
)
plt.ylabel('Count (log scale)')
plt.yscale('log')
plt.xticks(rotation=45)

# Format y-axis ticks with commas
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}' if x >= 1 else f'{x:.1e}'))

# Annotate each bar with its value
for p in ax.patches:
    value = int(p.get_height())
    if value > 0:
        ax.annotate(
            f'{value:,}',
            (p.get_x() + p.get_width() / 2, value),
            ha='center',
            va='bottom',
            fontsize=9,
            color='black',
            xytext=(0, 3),
            textcoords='offset points'
        )

plt.tight_layout()
plt.show()

# Display the number of unique countries and continents
# pd.set_option('display.max_rows', None)  # Show all rows
# display(country_continent.reset_index(drop=True))
# pd.reset_option('display.max_rows')      # Reset to default after


# Initail py codes. 
# # Bar plot: total cases and deaths by continent
# continent_summary = df_cleaned.groupby('continent')[['total_cases', 'total_deaths']].max().dropna()
# continent_summary.plot(kind='bar', figsize=(10, 6), title='Total Cases and Deaths by Continent')
# plt.ylabel('Count')
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

# ax = continent_summary.plot(kind='bar', figsize=(10, 6), title='Total Cases and Deaths by Continent')
# plt.ylabel('Count (log scale)')
# plt.yscale('log')
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

# import matplotlib.pyplot as plt
# import matplotlib.ticker as ticker

# ax = continent_summary.plot(
#     kind='bar',
#     figsize=(10, 6),
#     title='Total Cases and Deaths by Continent'
# )
# plt.ylabel('Count (log scale)')
# plt.yscale('log')
# plt.xticks(rotation=45)

# # Format y-axis ticks with commas
# ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}' if x >= 1 else f'{x:.1e}'))

# # Annotate each bar with its value
# for p in ax.patches:
#     value = int(p.get_height())
#     # Don't annotate if value is 0
#     if value > 0:
#         ax.annotate(
#             f'{value:,}',
#             (p.get_x() + p.get_width() / 2, value),
#             ha='center',
#             va='bottom',
#             fontsize=9,
#             color='black',
#             xytext=(0, 3),
#             textcoords='offset points'
#         )

# plt.tight_layout()
# plt.show()

# Calculate case fatality rate (CFR) as a percentage for each continent
continent_cfr = continent_summary.copy()
continent_cfr['CFR (%)'] = (continent_cfr['total_deaths'] / continent_cfr['total_cases']) * 100



```

### Explore the world wide monthly trend of COVID-19 cases using a line plot

```python
# Aggregate new cases monthly
monthly_trend = df_cleaned.groupby(['year', 'month'])['new_cases'].sum().reset_index()

# Create a 'date' column for plotting
monthly_trend['date'] = pd.to_datetime(monthly_trend[['year', 'month']].assign(day=1))

def millions(x, pos):
    # 'x' is the tick value; 'pos' is tick position (not used here)
    if x >= 1e6:
        return f'{x*1e-6:.0f}m'
    elif x >= 1e3:
        return f'{x*1e-3:.0f}k'
    else:
        return f'{x:.0f}'

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trend, x='date', y='new_cases')
plt.title('Worldwide Monthly COVID-19 Cases')
plt.ylabel('New Cases')
plt.xlabel('Date')
plt.xticks(rotation=45)

# Enable grid lines on the y-axis and x-axis
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, alpha=0.7)

ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(millions))

plt.tight_layout()
plt.show()
```

### Investigate the correlation between total cases and total deaths using a heatmap
